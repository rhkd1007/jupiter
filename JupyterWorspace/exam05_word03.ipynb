```python
import pandas as pd
import nltk # Natural Language Toolkit 자연어 처리 패키지
```

```python
nltk.download()
```

```python
all_data = pd.read_csv('riss_bigdata.csv')
all_title = all_data['제목']
all_data
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import re

stopWords=set(stopwords.words('english')) # NLTK가 정의한 영어 불용어 리스트
lemma = WordNetLemmatizer() # 표제어 추출 객체
words = []

for title in all_title:
#    print(title)
    enWords = re.sub(r'[^a-zA-Z]',' ',str(title)) #영어 아닌 것([^a-zA-Z]) '  ' 대체
    enWordsToken = word_tokenize(enWords.lower()) # 소문자로
#    print(enWordsToken)
    enWordsTokenStop = [w for w in enWordsToken if w not in stopWords ]
    enWordsTokenStopLemma = [lemma.lemmatize(w) for w in enWordsTokenStop]
#    print(enWordsTokenStopLemma)
    words.append(enWordsTokenStopLemma)
words
    
```

```python
# 전처리가 끝난 words 를 1차원 list
from functools import reduce
words2 = list(reduce(lambda x, y:x+y, words))
words2
```

```python
# 단어 빈도 구하기
from collections import Counter
count = Counter(words2)
count
```

```python
# 상위 50개 단어 딕셔너리에 저장
word_count = dict()
# count.most_common(50)
for tag, counts in count.most_common(50):
    if len(str(tag)) >1 :
        word_count[tag] = counts
word_count
```

```python
import matplotlib.pyplot as plt

sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='85')
plt.show()
```
