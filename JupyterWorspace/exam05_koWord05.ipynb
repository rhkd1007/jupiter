```python
import json
```

```python
# data = json.loads(open('4차 산업혁명.json','r',encoding='utf-8').read())
# data
f = open('4차 산업혁명.json','r',encoding='utf-8').read()
data = json.loads(f)
data
```

```python
import re
resultStr = ''
for item in data:
    # 문자와 숫자[A-a0-9_]== \w가 아닌것을 공백으로 치환
#     resultStr +=re.sub(r'[\w]',' ',item['message']) #문자와 숫자[A-a0-9_]== \w가 공백으로 치환
    resultStr += re.sub(r'[^\w]',' ',item['message']) # 문자와 숫자[A-a0-9_]== \w가 아닌것을 공백으로 치환
print(resultStr)
```

```python
message =''
for item in data:
    if'message' in item.keys():
        message = message + re.sub(r'[^\w]',' ',item['message'])
message
```

```python
from konlpy.tag import Okt # 한글 품사 태깅을 위한 모듈
```

```python
nlp = Okt()
message_N = nlp.nouns(resultStr) # nouns 명사 추출
message_N
```

```python
from collections import Counter
count = Counter(message_N)
count
```

```python
word_count = dict()
for tag, counts in count.most_common(80):
    if len(str(tag)) > 1 :
        word_count[tag] = counts
        print("%s : %d" %(tag,counts))
```

```python
#선생님 풀이
import matplotlib
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
```

```python
font_path = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc('font',family=font_name)
plt.figure(figsize=(12,5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)
sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='90')
plt.show()
```

```python
from wordcloud import WordCloud
wc = WordCloud(font_path, background_color='ivory', width=800, height=600)
cloud = wc.generate_from_frequencies(word_count)
plt.figure(figsize=(8,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
cloud.to_file('ko_wordCloud.jpg')
```
