```python
im = Image.open('ko_wordCloud.jpg')
size=(64,64)  # thumnail 
im.thumbnail(size)
im.save('ko_wordCloud_thum.jpg')
```

```python
im = Image.open('ko_wordCloud.jpg')
cropImage = im.crop((100,100,150,150))
cropImage.save('ko_wordCloud_crop.jpg')
```

```python
#90 
img3 = im.rotate(90)
img3.save('ko_wordCloud_ratate.jpg')
```

```python
im = Image.open('ko_wordCloud.jpg')
blurImage = im.filter(ImageFilter.BLUR)
blurImage.save('ko_wordCloud_blur.jpg')
```
